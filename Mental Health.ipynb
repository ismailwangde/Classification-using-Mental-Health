```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = nandi['Nbully'].values
y = nandi['Nage'].values
z = nandi['Nsatis'].values
x1 = nandi['Ibully'].values
y1 = nandi['Iage'].values
z1 = nandi['Isatis'].values

ax.scatter3D(x, y, z)
ax.scatter3D(x1, y1, z1)
#ax.scatter(1.5,2.5,3.5)
#plt.show()
import random as rn
import math, copy
netherl=[]
indl=[]
def rep(xrand,yrand):
  for i in range(10):
    netherl.append(math.sqrt(((xrand-nandi.loc[i][0])**2)+((yrand-nandi.loc[i][1])**2)+((zrand-nandi.loc[i][2])**2)))
    indl.append(math.sqrt(((xrand-nandi.loc[i][3])**2)+((yrand-nandi.loc[i][4])**2)+((zrand-nandi.loc[i][5])**2)))
  print(netherl,indl)
  n2,i2=copy.copy(netherl),copy.copy(indl)
  netherl.sort()
  indl.sort()
  for i in range(3):
    if((n2[n2.index(netherl[i])])<(i2[i2.index(indl[i])])):
      print("Netherland",n2[n2.index(netherl[i])]," at index ",n2.index(netherl[i]))
    else:
      print("India",i2[i2.index(indl[i])]," at index ",i2.index(indl[i]))

for i in range(1):
  xrand=rn.uniform(3.6,17.8)
  yrand=rn.uniform(0,43)
  zrand=rn.uniform(0.35,35.2)
  '''
  xrand=8.32
  yrand=18.9
  zrand=8.09'''
  print(xrand,yrand,zrand)
  ax.scatter3D(xrand,yrand,zrand)
  plt.show()
  rep(xrand,yrand)
  print(netherl,'xxxxxxxxxx',indl)
  print(netherl[0:3], indl[0:3])
```

```python
takeClosest = lambda num,collection:min(collection,key=lambda x:abs(x-num))
n=takeClosest(xrand/100,prob1["Nbully"].tolist())*takeClosest(yrand/100,prob1["Nage"].tolist())*takeClosest(zrand/100,prob1["Nsatis"].tolist())
neth=takeClosest(n,prob1.product(1)[0:9])
i=takeClosest(xrand/100,prob1["Ibully"][10:19].tolist())*takeClosest(yrand/100,prob1["Iage"][10:19].tolist())*takeClosest(zrand/100,prob1["Isatis"][10:19].tolist())
indi=takeClosest(i,prob1.product(1)[10:19])
if neth>indi:
  print("The new person belongs to Netherland")
else:
  print("The new person belongs to India")
```
